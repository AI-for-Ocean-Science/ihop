# Load up Loisel 23

# Examine one

```python
#idx = 0  # clear water
idx = 275 # high aph
#idx = 575 # 
#
a = ds.a.data[idx]
aph = ds.aph.data[idx]
ad = ds.ad.data[idx]
ag = ds.ag.data[idx]
anw = ds.anw.data[idx]
aw = a-anw
adg = ad+ag
wave = ds.Lambda.data
# bb
bb = ds.bb.data[idx]
bbd = ds.bbd.data[idx]
bbph = ds.bbph.data[idx]
bbnw = ds.bbnw.data[idx]
```

```python
## Plot
```

```python
fig = plt.figure(figsize=(7,7))
ax = plt.gca()

ax.plot(wave, a, 'k-', label='a')
ax.plot(wave, aph, 'g-', label='aph')
ax.plot(wave, adg, 'b-', label='adg')
ax.plot(wave, anw, 'r-', label='anw')
ax.plot(wave, aw, 'k:', label='aw')

ax.set_xlabel('Wave (nm)')
ax.set_ylabel(r'$a(\lambda)$')

ax.legend()

plt.show()
```

```python
fig = plt.figure(figsize=(7,7))
ax = plt.gca()

ax.plot(wave, bb, 'k-', label='bb')
ax.plot(wave, bbd, 'b-', label='bbd')
ax.plot(wave, bbph, 'g-', label='bbph')
ax.plot(wave, bbnw, 'r-', label='bbnw')
#ax.plot(wave, aw, 'k:', label='aw')

ax.set_xlabel('Wave (nm)')
ax.set_ylabel(r'$b_b(\lambda)$')

ax.legend()

plt.show()
```
