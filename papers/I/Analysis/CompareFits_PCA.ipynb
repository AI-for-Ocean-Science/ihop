# Explore RMSE

## $H_1$

```python
H1s = np.linspace(true_list[0]*0.99, true_list[0]*1.01, 100)
```

```python
modify = true_list.copy()
all_RMSE = []
for H1 in H1s:
    modify[0] = H1
    mod_Rs = emulator.prediction(modify, device)
    # RMSE
    rmse_mod = np.sum((mod_Rs-Rs[idx])**2)
    all_RMSE.append(rmse_mod)
```

## Plot

```python
fig = plt.figure(figsize=(7,7))
ax = plt.gca()

ax.plot(H1s, all_RMSE, 'o')
ax.axvline(true_list[0], color='k', ls='--')
#
#ax.legend()
plt.show()
```
