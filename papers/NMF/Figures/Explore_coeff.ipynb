# Explore coeff

```python
# imports
import numpy as np
import seaborn as sns
import pandas

import figs_nmf
```

```python
%matplotlib ipympl
```

# L23

```python
# load
nmf_fit = 'l23'
d = figs_nmf.load_nmf(nmf_fit)
M = d['M']
coeff = d['coeff']
wave = d['wave']
```

```python
df = pandas.DataFrame()
df['a0'] = coeff[:,0]
df['a1'] = coeff[:,1]
df['a2'] = coeff[:,2]
df['a3'] = coeff[:,3]
# logs
df['log_a0'] = np.log10(coeff[:,0])
df['log_a1'] = np.log10(coeff[:,1])
df['log_a2'] = np.log10(coeff[:,2])
df['log_a3'] = np.log10(coeff[:,3])
```

# Explore

```python
df.a0.min()
```

```python
sns.jointplot(data=df, x='log_a0', y='log_a1')
```

## $a_1$ vs. $a_2$

```python
sns.jointplot(data=df, x='log_a1', y='log_a2', color='green')
```
